# *Importing Modules*

```python
import pandas as pd
import seaborn as sns
```

# *Reading DF*

```python
df=pd.read_csv(r'D:\CSV_Files\housing.csv')
df.head()
```

# *Check for Skewed Data*

```python
sns.displot(df['RM'])
df['RM'].skew()
```

```python
sns.displot(df['LSTAT'])
df['LSTAT'].skew() #Right Skewed
```

```python
sns.displot(df['MEDV'])
df['MEDV'].skew() #almost Right Skewed
```

```python
sns.displot(df['PTRATIO'])
df['PTRATIO'].skew() #Left Skewed
```

# *Hint*

```python
# USE tranform if the model needed it: linear models
# Left Skewed ==> Square Transform
# Right ==> log transform
# check ==>  Yeo-Johnson 
df['new_PTRATIO']=df['PTRATIO'].apply(lambda x : x**2)
sns.displot(df['new_PTRATIO'])
df['new_PTRATIO'].skew() #Left Skewed
```

```python
from sklearn.preprocessing import PowerTransformer
transform=PowerTransformer(standardize=True)
df['new_PTRATIO']=transform.fit_transform(df['PTRATIO'].values.reshape(-1,1))
sns.distplot(df['new_PTRATIO'])
df['new_PTRATIO'].skew()
```

```python
import numpy as np
df['new_MEDV']=df['MEDV'].apply(lambda x : np.log(x))
sns.displot(df['new_MEDV'])
df['new_MEDV'].skew() #almost Right Skewed
```

```python
transform=PowerTransformer(standardize=True)
df['new_MEDV']=transform.fit_transform(df['MEDV'].values.reshape(-1,1))
sns.distplot(df['new_MEDV'])
df['new_MEDV'].skew()
```
