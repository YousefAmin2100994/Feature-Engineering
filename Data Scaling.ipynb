# *Importing Modules*

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.preprocessing import StandardScaler,MaxAbsScaler,MinMaxScaler
```

## *Reading DataFrame*

```python
data_frame=pd.read_csv(r'H:\StudentsPerformance.csv')
data_frame.head()
```

## *1.StandardScaler*

```python
scale=StandardScaler() # x=(x-mean)/std , used if the data is normal direct 
x=scale.fit_transform(data_frame[['reading score','writing score']])
x
```

```python
pd.DataFrame(x,columns=['reading score','writing score'])
```

## *2.MinMaxScaler*

```python
scale=MinMaxScaler() # x = (x-mean)/range , used if there are alot of outliers
x=scale.fit_transform(data_frame[['reading score','writing score']])
x
```

```python
pd.DataFrame(x,columns=['reading score','writing score'])
```

## *3.MaxAbs*

```python
scale=MaxAbsScaler() # x = x/max , used if there are alot of zeros in data 
x=scale.fit_transform(data_frame[['reading score','writing score']])
x
```

```python
pd.DataFrame(x,columns=['reading score','writing score'])
```

## *4.FunctionTransformer*

```python
from sklearn.preprocessing import FunctionTransformer
scale=FunctionTransformer(func=lambda x : x**0.1)
x=scale.fit_transform(data_frame[['reading score','writing score']])
x
```
