# *Pie Plot*

```python
df=pd.read_csv(r'D:\ITI-Emy Raslan\EDA\Matplotlib\pie_chart_data.csv')
df.head()
```

```python
plt.pie(df['Number of Cars'],labels=df['Engine Fuel Type'],autopct='%1.1f%%',explode=[0.02,0.02,0.02,0.02])
plt.title('Cars by Engine Fuel Type ', fontsize=16,fontweight='bold')
```

```python
#Pie plots are useful for illustrating how individual parts contribute to a whole, 
#particularly when dealing with a small number of categories and clear proportional differences. 
#However, they should be used with caution, as they can become misleading or hard to interpret when there are too many categories or when the differences between categories are subtle.
```

# *Line Plot*

```python
df=pd.read_csv(r'D:\ITI-Emy Raslan\EDA\Matplotlib\line_chart_data.csv')
df.head()
```

```python
df.info() #notice that the Date is obj 
```

```python
df['new_date']=pd.to_datetime(df['Date'])
df.head()
```

```python
df.info()
```

```python
df['Month']=df['new_date'].apply(lambda x : x.month)
df['Day']=df["new_date"].apply(lambda x:x.day)
df['Year']=df['new_date'].apply(lambda x:x.year)
df.head()
```

```python
plt.figure(figsize=(20,6))
plt.plot(df['new_date'],df['GSPC500'])
plt.xlabel("year")
plt.ylabel("GSPC500")
```

```python
plt.figure(figsize=(20,6))
new_frame=df[df['Year']==2010]
plt.plot(new_frame['new_date'],new_frame['GSPC500'])
```

# *Histogram*

```python
df=pd.read_csv(r'D:\ITI-Emy Raslan\EDA\Matplotlib\histogram_data.csv')
df.head()
```

```python
hist, bin_edges = np.histogram(df['Price'], bins='auto')
plt.hist(df['Price'],bins=bin_edges, color='skyblue', edgecolor='black', alpha=0.7)
plt.ylabel("Freq")
plt.xlabel("Price")
plt.title("Distribution")
```

```python
sns.kdeplot(df['Price'])
```
