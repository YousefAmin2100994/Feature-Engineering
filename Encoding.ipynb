# *Importing Modules*

```python
import pandas as pd
import numpy as np
```

# *Reading Df*

```python
df=pd.read_csv(r'Downloads\WA_Fn-UseC_-HR-Employee-Attrition.csv')
df.head()
```

# *Label Encoding*

```python
df['Attrition']=df['Attrition'].apply(lambda x : 1 if x=='Yes' else 0)
df.head()
```

# *One Hot Encoding*

```python
df=pd.get_dummies(df,columns=list(df.select_dtypes(include='object').columns),drop_first=True)
df.head()
```
